```python
df=df.dropna(subset=["CustomerID"])
```

```python
df["CustomerID"]=df["CustomerID"].astype(int)
```

```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
```

```python
df["is cancelled"]=df["InvoiceNo"].astype(str).str.startswith("c")
```

```python
non_pruduct_codes=["POST","BANK CHARGES","ADJUSTMENT","DOT"]
```

```python
df[~df["StockCode"].isin(non_pruduct_codes)]
```

```python
df=df[df["UnitPrice"]>0]
df=df[df["Quantity"]!=0]
```

```python
df["Description"]=df["Description"].fillna("Unknown Product")
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
print("final sahape:",df.shape)
print(df.head())
```

```python
df["is cancelled"].value_counts()
```

```python
reference_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)
```

```python
customer_df = (
    df.groupby("CustomerID")
    .agg(
        total_revenue=("Revenue", "sum"),       # مجموع درآمد
        total_orders=("InvoiceNo", "nunique"),  # تعداد فاکتور
        total_quantity=("Quantity", "sum"),     # تعداد کل کالا خریداری شده
        last_purchase=("InvoiceDate", "max"),   # آخرین خرید
        total_returns=("is cancelled", "sum")    # تعداد مرجوعی‌ها
    )
    .reset_index()
)
```

```python
customer_df["recency_days"] = (
    reference_date - customer_df["last_purchase"]
).dt.days
```

```python
customer_df["return_rate"] = (
    customer_df["total_returns"] / customer_df["total_orders"]
)
```

```python
customer_df = customer_df[customer_df["total_revenue"] > 0]

customer_df = customer_df.sort_values(by="total_revenue", ascending=False)
```

```python
customer_df.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

customer_df[["revenue_score", "recency_score", "return_score"]] = scaler.fit_transform(
    customer_df[["total_revenue", "recency_days", "return_rate"]]
)

customer_df["recency_score"] = 1 - customer_df["recency_score"]
```

```python
customer_df["decision_score"] = (
    0.5 * customer_df["revenue_score"] +
    0.3 * customer_df["recency_score"] -
    0.2 * customer_df["return_score"]
)
```

```python
def label_customer(score):
    if score >= 0.6:
        return "Keep"
    elif score >= 0.4:
        return "Grow"
    else:
        return "Drop"

customer_df["decision_label"] = customer_df["decision_score"].apply(label_customer)
```

```python
print(customer_df["decision_label"].value_counts())

print(customer_df.sort_values("decision_score", ascending=False).head(10))
```

```python
final_cols = [
    "CustomerID",
    "total_revenue",
    "total_orders",
    "recency_days",
    "return_rate",
    "decision_score",
    "decision_label"
]

final_df = customer_df[final_cols]

final_df.to_csv("customer_decision_pipeline_final.csv", index=False)
```
